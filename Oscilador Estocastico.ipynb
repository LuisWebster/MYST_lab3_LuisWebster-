# Oscilador Estocastico

Esta técnica para detectar señales de compra y venta fue desarrollada durante los años 50s por George C. Lane. Cabe destacar que este indicador no sigue el precio, volumen etc. sino que le da seguimiento a la velocidad del precio para un rango de tiempo determinado. Podemos decir (como regla general) que el impulso cambio de dirección antes de que lo haga el precio. 

Este oscilador permite identificar condiciones alcistas o bajistas del mercado, lo cual es de gran utilidad para que los inversionistas conozcan futuras (potenciales) reversiones en los mercados financieros.

%$K= (CierreActual-MínimoMásBajo)/(MáximoMásAlto-MínimoMásBbajo) x 100 $

%$D$ = Media móvil simple de 3 periodos del %K

Sabemos por default que el oscilador estocástico consta de 14 periodos, los cuales pueden ser horas, semanas, días, meses, etc. Esto dependerá totalmente del usuario. A continuación se presentan las interpretaciones que se le da a este indicador:

* El oscilador estará arriba de 50 cuando el precio de cierre esté en la "mitad de arriba"
* El oscilador estará debajo de 50 cuando el precio de cierre esté en la "mitad de abajo"
* Valores superiores a 80 indican que el precio de cierre está cerca del máximo durante el periodo de tiempo estudiado
* Valores inferiores a 20 indican que el precio de cierre está cerca del mínimo durante el periodo de tiempo estudiado

Sabemos también que existen 3 tipos de osciladores, los cuales si diferencian (principalmente) por su velocidad:

* Lento: se calculará la %K con un SMA (simple moving average) de 3 periodos, para suavizar los datos
* Rápido: calculo regular de %k (descrito arriba) 
* Completo: el usuario escogerá cuando periodos desea tomar en cuenta para el cálculo de SMA
